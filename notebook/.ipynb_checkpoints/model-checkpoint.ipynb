```python
def date_generator(begin_date, end_date):
    begin =datetime.datetime.strptime(begin_date, "%Y%m%d")
    end = datetime.datetime.strptime(end_date, "%Y%m%d")
    return [(begin + datetime.timedelta(days=x)).strftime("%Y-%m-%d") for x in range(0, (end-begin).days + 1)]
```

#### 2. 속성 생성

```python
def fill_missing_values(list_val):
    list_tmp = []
    for v in list_val:
        if not math.isnan(v) :
            prev_v = v
        list_tmp.append(prev_v)
    return list_tmp
```

```python
df_usd_krw = df_usd_krw[['Close']].rename(columns={'Close':'y'})
df_kospi = df_kospi[['Close']].rename(columns={'Close':'kospi_close'})
```

```python
df_us_bond = df_bond1y.merge(df_bond10y, on='Date')
df_us_bond['bond_rate_diff'] = df_us_bond['10y_close'] - df_us_bond['1y_close']
df_us_bond = df_us_bond[['Date', 'bond_rate_diff']]
df_us_bond['Date'] = pd.to_datetime(df_us_bond['Date'])
df_us_bond.set_index('Date', inplace=True)


df_bb_rated = pd.read_csv('BBrated.csv')
df_bb_rated.columns = ['Date', 'bb_rated']
df_bb_rated['Date'] = pd.to_datetime(df_bb_rated['Date'])
df_bb_rated.loc[(df_bb_rated.bb_rated == '.'), 'bb_rated'] = np.nan
df_bb_rated['bb_rated'] = df_bb_rated['bb_rated'].astype(float)
df_bb_rated.set_index('Date', inplace=True)
```

```python
df_oil['Date'] = pd.to_datetime(df_oil['Date'])
df_oil.set_index('Date', inplace=True)
df_oil.head()
```

```python
df_data = df_usd_krw.join(df_kospi, how='left').join(df_us_bond, how='left').join(df_bb_rated, how='left').join(df_naver, how='left').join(df_oil,  how='left')
df_data = df_data[df_data.index >= '2019-04-13']
df_data.head()
```

```python
for col in df_data.columns:
    df_data[col] =  fill_missing_values(df_data[col].values)
```

```python
poly = PolynomialFeatures(degree=4, include_bias=True)
scaler = StandardScaler()

X = df_data.values[:, 1:6]
y = df_data.values[:, 0]

# 훈련 데이터셋 X_train 의 거듭제곱을 생성한 뒤, 훈련 데이터셋 X_train 에 새로운 변수로 추가
X_scaled = scaler.fit_transform(X)
X_poly = poly.fit_transform(X_scaled)


X_train = X_poly[:int(len(X_poly) - 3)]
X_test = X_poly[int(len(X_poly) - 3):]
y_train = y[:int(len(X_poly) - 3)]
y_test = y[int(len(X_poly) - 3):]


lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_reg.score(X_train, y_train)
```
